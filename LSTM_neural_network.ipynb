```python
def df_to_X_y(df, window_size=50):
    # Convert the DataFrame to a numpy array
    df_as_np = df.to_numpy()
    
    # Initialize lists for input features and target labels
    X = []
    y = []
    
    # Loop through the rows of the numpy array
    for i in range(len(df_as_np)-window_size):
        # Extract a window of rows with length 'window_size'
        row = [r for r in df_as_np[i:i+window_size]]
        
        # Append the window of rows to the list of input features
        X.append(row)
        
        # Extract the last two columns of the last row in the window as the target labels
        label = df_as_np[i+window_size][-2:]
        
        # Append the target labels to the list of target labels
        y.append(label)
    
    # Convert the input features and target labels to numpy arrays
    X = np.array(X)
    y = np.array(y)
    
    return X, y
```

```python
from keras.models import Sequential
from keras.layers import *
from keras.losses import MeanSquaredError
from keras.metrics import RootMeanSquaredError
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(InputLayer((50, 6)))
model.add(LSTM(64,activation='sigmoid'))
model.add(Dense(8, 'relu'))
model.add(Dense(2, 'linear'))
model.summary()
```

```python
X_train = np.asarray(X_train).astype(np.float32)
y_train = np.asarray(y_train).astype(np.float32)
X_val = np.asarray(X_val).astype(np.float32)
y_val = np.asarray(y_val).astype(np.float32)
```

```python
X_test = np.asarray(X_test).astype(np.float32)
y_test = np.asarray(y_test).astype(np.float32)
```

```python
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
m_checkpoint = ModelCheckpoint('model_lstm/', save_best_only=True)
```

```python
model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.00000001), metrics=[RootMeanSquaredError()])
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=6)
```
